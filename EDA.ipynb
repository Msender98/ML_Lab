```python
import pandas as pd
```

```python
orders_df = pd.read_csv('./data/Orders.csv')
```

```python
orders_df.shape
orders_df.head()
```

```python
def money(price):
    return(float(price.replace('$','').replace(',','')))
```

```python
orders_df.loc[:,'Profit'] = orders_df.loc[:,'Profit'].apply(money)
```

```python
orders_df.loc[:,'Sales'] = orders_df.loc[:,'Sales'].apply(money)
```

```python
orders_df.loc[:,'Order.Date'] = orders_df.loc[:,'Order.Date'].apply(pd.to_datetime)
```

```python
orders_df['Quarter'] = orders_df['Order.Date'].apply(lambda x: (x.month-1)//3)
```

```python
orders_df.groupby('Quarter').aggregate(sum)
```
